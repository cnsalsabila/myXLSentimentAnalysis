```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from collections import Counter
import re
import pandas as pd
```

```python
import io

filename = list(uploaded.keys())[0]
df_myxl_rev = pd.read_csv(io.BytesIO(uploaded[filename]))

print(f'Shape dataset: {df_myxl_rev.shape}')
df_myxl_rev.head()
```

```python
# Ganti seluruh cell itu dengan ini saja
corpus = df_myxl_rev['processed_text'].dropna().tolist()
corpus = [s for s in corpus if str(s).strip() != '']

print(f'Total dokumen dalam corpus: {len(corpus)}')
print('\nContoh 5 dokumen pertama:')
print(corpus[:5])
```

## List all unique words

```python
print("Corpus (5 sampel pertama):", corpus[:5])
```

```python
# Finding unique words
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print(f'Total unique words: {len(unique_words)}')
print('Contoh 20 unique words:', unique_words[:20])
```

## Create a dictionary with mapping of words to a number

```python
all_words = []
for sentence in corpus:
    for word in sentence.split():
        all_words.append(word)

word_frequencies = Counter(all_words)

# Print top 20 word counts (karena dataset besar)
print('Top 20 Word Frequencies:', word_frequencies.most_common(20))

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print(f'\nTotal vocabulary size: {len(word_to_number)}')
print('Word to Number Mapping (Top 20):', dict(list(word_to_number.items())[:20]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus_sample = corpus[:100000]

cv = CountVectorizer(max_features=2000)
independentFeatures = cv.fit_transform(corpus_sample).toarray()

print(f'Shape BoW matrix: {independentFeatures.shape}')
print('(jumlah dokumen x jumlah fitur/vocabulary)')
print('\nContoh BoW matrix (5 dokumen pertama, 10 fitur pertama):')
print(independentFeatures[:5, :10])
```

```python
# Tampilkan BoW sebagai DataFrame agar lebih mudah dibaca
bow_df = pd.DataFrame(independentFeatures, columns=cv.get_feature_names_out())
print('BoW DataFrame (5 baris pertama):')
bow_df.head()
```

```python
# Simpan BoW matrix ke CSV
bow_df.to_csv('df_myxl_bow.csv', index=False)
print('BoW matrix berhasil disimpan sebagai df_myxl_bow.csv')

files.download('df_myxl_bow.csv')
```
