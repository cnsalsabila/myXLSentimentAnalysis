```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (Artikel 1: Pemakaman Komandan)
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan Hasil
df = pd.DataFrame(X.toarray(), columns=['iran', 'serangan', 'pemimpin', 'khamenei', 'israel', 'warga', 'teheran'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata-kata kunci seperti `iran`, `serangan`, `khamenei`, dll. agar dapat memvisualisasikan kedekatan topik antar artikel.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Ketiga artikel membahas topik yang sama (situasi Iran), sehingga diharapkan memiliki nilai similarity yang tinggi satu sama lain — berbeda dengan corpus Stanford yang memiliki topik lebih bervariasi.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, artikel yang lebih panjang dengan banyak pengulangan kata kunci akan dianggap sangat berbeda dari artikel pendek — padahal topiknya sama.

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
